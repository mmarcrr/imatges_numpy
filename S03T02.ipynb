### Exercici 4
Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
#mostra broadcast matrius 2D
mymatrix_2DA=np.random.randint(1,99, size=(5,4), dtype=np.int64)
mymatrix_2DB=np.random.randint(1,99, size=(5,4), dtype=np.int64)
print(mymatrix_2DA+mymatrix_2DB)
print(mymatrix_2DA-mymatrix_2DB)
print(mymatrix_2DA*mymatrix_2DB)
print(mymatrix_2DA/mymatrix_2DB)
print(mymatrix_2DA%mymatrix_2DB)
print(mymatrix_2DA==mymatrix_2DB)
```

```python
#mostra boradcas matrius 4D
mymatrix_4DA=np.random.randint(1,99, size=(5,4,3,2), dtype=np.int64)
mymatrix_4DB=np.random.randint(1,99, size=(5,4,3,2), dtype=np.int64)
print(mymatrix_4DA+mymatrix_4DB)
print(mymatrix_4DA-mymatrix_4DB)
print(mymatrix_4DA*mymatrix_4DB)
print(mymatrix_4DA/mymatrix_4DB)
print(mymatrix_4DA%mymatrix_4DB)
print(mymatrix_4DA==mymatrix_4DB)
```

```python
#matriu dimensio 1 am matriu dimensio 4
mymatrix_1DA=np.random.randint(1,99, size=(1), dtype=np.int64)
print(mymatrix_1DA+mymatrix_4DB)
print(mymatrix_1DA-mymatrix_4DB)
print(mymatrix_1DA*mymatrix_4DB)
print(mymatrix_1DA/mymatrix_4DB)
print(mymatrix_1DA%mymatrix_4DB)
print(mymatrix_1DA==mymatrix_4DB)
```

```python
#mostra broadcast matrius de diferent dimensio
mymatrix_2DC=np.random.randint(1,99, size=(5,3), dtype=np.int64)
try:
    print(mymatrix_2DA+mymatrix_2DC)
except Exception as e:
    print(e)


```

```python
##en aquest cas ens dona False doan a entendr que les matrius son de diferent dimensio
try:
    print(mymatrix_2DA==mymatrix_2DC)
except Exception as e:
    print(e)
```
